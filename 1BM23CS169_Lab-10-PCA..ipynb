```
# ================================
# STEP 1: Import Libraries
# ================================
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score


# ================================
# STEP 2: Load Dataset
# ================================
data = pd.read_csv("heart.csv")

# Clean column names (good practice)
data.columns = data.columns.str.strip()

print("Columns:", data.columns)


# ================================
# STEP 3: Encode Categorical Data
# ================================
# Convert text → numbers
data = pd.get_dummies(data, drop_first=True)

print("\nAfter Encoding:")
print(data.head())


# ================================
# STEP 4: Split Features & Target
# ================================
X = data.drop("HeartDisease", axis=1)
y = data["HeartDisease"]


# ================================
# STEP 5: Scaling
# ================================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# ================================
# STEP 6: Train-Test Split
# ================================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)


# ================================
# STEP 7: Models BEFORE PCA
# ================================

# Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
acc_lr = accuracy_score(y_test, lr.predict(X_test))

# SVM
svm = SVC()
svm.fit(X_train, y_train)
acc_svm = accuracy_score(y_test, svm.predict(X_test))

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
acc_rf = accuracy_score(y_test, rf.predict(X_test))


print("\n===== BEFORE PCA =====")
print("Logistic Regression Accuracy:", acc_lr)
print("SVM Accuracy:", acc_svm)
print("Random Forest Accuracy:", acc_rf)

best_before = max(acc_lr, acc_svm, acc_rf)
print("Best Accuracy BEFORE PCA:", best_before)


# ================================
# STEP 8: Apply PCA
# ================================
pca = PCA(n_components=5)   # try 2, 3, 5 if needed
X_pca = pca.fit_transform(X_scaled)

print("\nExplained Variance Ratio:", pca.explained_variance_ratio_)


# ================================
# STEP 9: Train-Test Split (PCA)
# ================================
X_train_pca, X_test_pca, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.2, random_state=42
)


# ================================
# STEP 10: Models AFTER PCA
# ================================

# Logistic Regression
lr.fit(X_train_pca, y_train)
acc_lr_pca = accuracy_score(y_test, lr.predict(X_test_pca))

# SVM
svm.fit(X_train_pca, y_train)
acc_svm_pca = accuracy_score(y_test, svm.predict(X_test_pca))

# Random Forest
rf.fit(X_train_pca, y_train)
acc_rf_pca = accuracy_score(y_test, rf.predict(X_test_pca))


print("\n===== AFTER PCA =====")
print("Logistic Regression Accuracy:", acc_lr_pca)
print("SVM Accuracy:", acc_svm_pca)
print("Random Forest Accuracy:", acc_rf_pca)

best_after = max(acc_lr_pca, acc_svm_pca, acc_rf_pca)
print("Best Accuracy AFTER PCA:", best_after)
```
